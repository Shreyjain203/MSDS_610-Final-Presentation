```python
# Import Libraries
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

# Agenda: Compare the Mean Values of Two groups
### HO: Mean_GroupA = Mean_GroupB
### H1: Mean_GroupA != Mean_GroupB

#### Check the rejection by using T-test and p-value

```python
# Simulate data for two groups (A and B)
np.random.seed(42)  # Set seed for reproducibility

# Generate data for control group (Group A)
group_a = np.random.normal(loc=24, scale=5, size=1000)

# Generate data for treatment group (Group B) with a slight increase
group_b = np.random.normal(loc=29, scale=5, size=1000)

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(group_a, group_b)

# Check the p-value against the significance level
alpha = 0.05

if p_value < alpha:
    print(f"P-value is {p_value:.5f} which is less than alpha. Reject the null hypothesis. There is a significant difference between the groups.")
else:
    print(f"P-value is {p_value:.5f} which is greater than alpha. Fail to reject the null hypothesis. There is no significant difference between the groups.")
```

#### Draw the histogram and see the difference

```python
# Function to generate and plot sample data
def generate_and_plot_samples(group_a, group_b, xlabel, title):
    plt.hist(group_a, alpha=0.5, label='Group A')
    plt.hist(group_b, alpha=0.5, label='Group B')
    plt.legend()
    plt.xlabel(xlabel)
    plt.title(title)
    plt.show()
    
    
# Scenario for Type II Error (Null Hypothesis is False)
group_a = np.random.normal(loc=24, scale=5, size=1000)
group_b = np.random.normal(loc=29, scale=5, size=1000)
generate_and_plot_samples(group_a, group_b,
                          "Value", "Two Means by Group")
```

Two groups have different mean values so reject the null

```python
import numpy as np
import statsmodels.api as sm

# Set a random seed for reproducibility
np.random.seed(42)

# Function to add outliers to a dataset
def add_outliers(data, num_outliers, loc=300, scale=30):
    outliers = np.random.normal(loc=loc, scale=scale, size=num_outliers)
    return np.concatenate([data, outliers])

# Scenario for Type I Error (Null Hypothesis is True)
group_a_typeI = np.random.normal(loc=25, scale=5, size=1000)
group_b_typeI = np.random.normal(loc=25, scale=5, size=1000)

# Add outliers to Scenario Type I
group_a_typeI = add_outliers(group_a_typeI, num_outliers=50)
group_b_typeI = add_outliers(group_b_typeI, num_outliers=50)

# Perform two-sample t-test using statsmodels
t_stat_typeI, p_value_typeI, _ = sm.stats.ttest_ind(group_a_typeI, group_b_typeI)

print(p_value_typeI)
# Check the p-value against the significance level
alpha = 0.05
if p_value_typeI < alpha:
    print("Scenario Type I - Reject the null hypothesis. Incorrectly concluded a significant difference (Type I Error).")
else:
    print("Scenario Type I - Fail to reject the null hypothesis. Correct decision. No significant difference.")

# Scenario for Type II Error (Null Hypothesis is False)
group_a_typeII = np.random.normal(loc=24, scale=5, size=1000)
group_b_typeII = np.random.normal(loc=29, scale=5, size=1000)  # Introduce a difference

# Add outliers to Scenario Type II
group_a_typeII = add_outliers(group_a_typeII, num_outliers=50)
group_b_typeII = add_outliers(group_b_typeII, num_outliers=50)

# Perform two-sample t-test using statsmodels
t_stat_typeII, p_value_typeII, _ = sm.stats.ttest_ind(group_a_typeII, group_b_typeII)
print(p_value_typeII)
# Check the p-value against the significance level
if p_value_typeII < alpha:
    print("Scenario Type II - Reject the null hypothesis. Correct decision. A significant difference exists.")
else:
    print("Scenario Type II - Fail to reject the null hypothesis. Type II Error. Failed to detect a significant difference.")
```
